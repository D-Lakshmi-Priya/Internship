```python
townName = []
matchesPlayed = []
TownData = df['home_team'].value_counts()
for _ in TownData.index:
  if TownData[_] > 30:
    townName.append(_)
    matchesPlayed.append(TownData[_])
```

```python
townName
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(matchesPlayed,townName)
plt.xlabel('Town Name')
plt.ylabel('Matches Played')
plt.title('Matches Played by Town')
plt.show()
```

```python
plt.bar(matchesPlayed,townName)
plt.ylabel('Town Name')
plt.xlabel('Matches Played')
plt.title('Matches Played by Town')
plt.show()
```

```python
def MatchesPlayedByTown():
  townName = []
  matchesPlayed = []
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    if TownData[_] > 30:
      townName.append(_)
      matchesPlayed.append(TownData[_])
  plt.bar(matchesPlayed,townName)
  plt.ylabel('Town Name')
  plt.xlabel('Matches Played')
  plt.title('Matches Played by Towns')
  plt.show()
```

```python
MatchesPlayedByTown()
```
