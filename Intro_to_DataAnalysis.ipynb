```python
import random
```

```python
a1=[]
a2=[]
a3=[]
a4=[]
a5=[]
```

```python
for _ in range(1000):
    a1.append(random.randint(0,100))
    a2.append(random.randint(50,70))
    a3.append(random.randint(10,20))
    a4.append(random.randint(5,10))
    a5.append(random.randint(20,100))
```

```python
df = {
     'grade':a1,
     'weight':a2,
     'friends':a3,
     'cousins':a4,
     'projects':a5
}
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(df)
df1
```

```python
df1.describe()
```

```python
df1['grade'].count()
```

```python
freind15 = df1['friends']>15
df1[freind15]
```

```python
grade75 = df1['grade']>75
df1[grade75]
```
