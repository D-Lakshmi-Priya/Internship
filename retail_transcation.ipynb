```python
# Task3-Data Cleaning:
# Checking for missing values in each column and printing the count per column
df.isnull().sum()
```

```python
#for the columns Quantity, Price, and Discount, to ensure they are in the correct numrical formate
df['Quantity']=pd.to_numeric(df['Quantity'])
df['Price']=pd.to_numeric(df['Price'])
df['Discount']=pd.to_numeric(df['Discount'])

df['Discount']= df['Discount'].fillna(0)
```

```python
df
```

```python
# Task4-Feature Engineering:
# Creating a new column called NetPrice which applies the discount to the price.
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)
df['NetPrice']
```

```python
# Creating new column called TotalSale
df['TotalSale'] = df['Quantity'] * df['NetPrice']
df['TotalSale']
```

```python
# Task5-Data Filtering:
# Filtering the DataFrame to include only those transaction where TotalSale is greater than or equal to 20
filtered_df = df[df['TotalSale'] >= 20]
filtered_df
```

```python
# Saving filtered data into a new DataFrame for the futther analysis
df1 = filtered_df
```
